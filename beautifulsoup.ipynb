```python
from bs4 import BeautifulSoup
import requests
url='https://www.amazon.in/s?k=laptops&crid=8NGFB4RDIK1A&sprefix=laptops%2Caps%2C200&ref=nb_sb_noss_2'
```

```python
res=requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
h2_tag=soup.find_all('h2',class_='a-size-medium a-spacing-none a-color-base a-text-normal')
h2_tag
```

```python
span_list=[]
for span in h2_tag:
    span_list.append(span.find('span').string)
print(span_list)
```

```python
price=soup.find_all('span',class_="a-price-whole")
```

```python
price_list=[]
for j in price:
    price_list.append(j.string)
price_list
```

```python
import pandas as pd
data={'laptop':span_list,'prices':price_list}
df=pd.DataFrame(data)
df
```

```python
df.to_csv('beautiful.csv',index=False)
```
